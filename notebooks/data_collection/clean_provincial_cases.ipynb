```python
import pandas as pd
```

```python
cases = pd.read_csv('../../data/covid19za_provincial_cumulative_timeline_confirmed.csv')
cases.shape
```

```python
cases.head(10)
```

```python
ec = cases[['date', 'YYYYMMDD', 'EC']]
gp = cases[['date', 'YYYYMMDD', 'GP']]
fs = cases[['date', 'YYYYMMDD', 'FS']]
kzn = cases[['date', 'YYYYMMDD', 'KZN']]
lp = cases[['date', 'YYYYMMDD', 'LP']]
mp = cases[['date', 'YYYYMMDD', 'MP']]
nc = cases[['date', 'YYYYMMDD', 'NC']]
nw = cases[['date', 'YYYYMMDD', 'NW']]
wc = cases[['date', 'YYYYMMDD', "WC"]]
```

```python
p = ['EC', 'GP', 'FS', 'KZN', 'LP', 'MP', 'LP', 'NC', 'NW', 'WC']
```

```python
dataframes = []

for province in p:
    df = cases[['date', 'YYYYMMDD', province.upper()]]
    df = df.rename(columns={province.upper(): 'Cases'})
    df = df.loc[df['Cases'] != 0]
    df['Province'] = province.upper()
    dataframes.append(df)

all_data = pd.concat(dataframes)

```

```python
all_data.sample(n=50)
```

```python
all_data.to_csv('../../data/cleaned/provincial_cases.csv')
```
