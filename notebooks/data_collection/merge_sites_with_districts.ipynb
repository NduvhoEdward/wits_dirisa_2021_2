```python
from fuzzywuzzy import process, fuzz
```

```python
unique_muni = muni["MUNICNAME"].unique()
```

```python
process.extract('Khayelitsha', unique_muni, scorer=fuzz.token_set_ratio)
```

This method of linking seems to be verry promissing:

One thing that has been encounterd is that the sites data sub-district ends with LM (local municipality) or Municipality wich drastically lowers the score. However the token_set_ratio seems to really out perform the token_sort_ratio a threshold fo 75 should be tested

```python
sub_district_map = {}

poor_results = []


for index, row in sites_e.iterrows():
    search = process.extract(row['Sub-District'], unique_muni, scorer=fuzz.token_set_ratio)

    if len(search) > 0 and search[0][1] >= 65:
        sub_district_map[search[0][0]] = row['Sub-District']
    else:
        poor_results.append({'Sub-District': row['Sub-District'], 'search': search})


print('Subdistricts Identified: ', len(sub_district_map))
print(sub_district_map)
    
```

```python
# It seems that a good portion of subdistricts have been mapped however there are still quite a few that don't have a 
# mapping, the best could be to map the districts.
```

```python
# But first add the one's that are well classified
s = set( val for dic in sub_district_map for val in sub_district_map.values())
len(s)

reversed_sub_dist_map = {value : key for (key, value) in sub_district_map.items()}

sites_ev2 = gpd.GeoDataFrame(sites_e)

sites_ev2['Municipality'] = sites_e["Sub-District"].map(reversed_sub_dist_map)

sites_ev2.sample(n=20)


```

```python
# Count no of nan rows

percent_missing = sites_ev2.isna().sum() * 100 / len(sites_ev2)
missing_value_df = pd.DataFrame({"col": sites_ev2.columns, 'Percent Missing': percent_missing})
missing_value_df
```

```python
unique_districts = muni["DISTRICT_N"].unique()
len(unique_districts)
```

```python
search = process.extract('Amathole DM', unique_districts, scorer=fuzz.token_set_ratio)
search
```

```python
district_map = {}

poor_results2 = []


for index, row in sites_e.iterrows():
    search = process.extract(row['District'], unique_districts, scorer=fuzz.token_set_ratio)

    if len(search) > 0 and search[0][1] >= 63:
        district_map[search[0][0]] = row['District']
    else:
        poor_results2.append({'District': row['District'], 'search': search})


print('Subdistricts Identified: ', len(district_map))
print(district_map)
print(len(poor_results2))
```

```python
s = set( val for dic in district_map for val in district_map.values())
print(len(s))

reversed_dist_map = {value : key for (key, value) in district_map.items()}

reversed_dist_map['Central Karoo'] = 'Central Karoo'
reversed_dist_map['Garden Route'] = 'Garden Route'

sites_ev3 = gpd.GeoDataFrame(sites_ev2)

sites_ev3['DISTRICT_N'] = sites_e["District"].map(reversed_dist_map)
sites_ev3['Municipality'] = sites_e["Sub-District"].map(reversed_sub_dist_map)
sites_ev3.sample(n=20)
```

```python
# Count no of nan rows
def getCompleteness(df):
    percent_missing2 = df.isna().sum() * 100 / len(df)
    missing_value_df2 = pd.DataFrame({"col": df.columns, 'Percent Missing': percent_missing2})
    return missing_value_df2
```
