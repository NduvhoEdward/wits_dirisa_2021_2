The data can be read quite easily. 

```python
provinces = ['eastern_cape', 'western_cape', 'gauteng', 'northen_cape', 'north_west', 'free_state', 'kwazulu_natal', 'mpumalanga', 'limpopo']
len(provinces)
```

```python
from os import path

dataset_loc = "../../data/vaccination-sites"

dataframes = []

for province in provinces:
    loc = path.join(dataset_loc, (province+".xlsx"))
    temp = pd.read_excel(loc)
    dataframes.append(temp)
```

```python
vaccination_sites = pd.concat(dataframes)
print(vaccination_sites.shape)
vaccination_sites
```

```python
vaccination_sites.to_csv('../../data/cleaned/sacoronavirus_vaccination_sites.csv')
```
