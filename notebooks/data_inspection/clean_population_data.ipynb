```python
import pandas as pd
```

```python
sa_districts = pd.read_csv("../../data/cleaned/sa_districts.csv")
print(sa_districts.shape)
sa_districts.sample(n=5)
```

```python
pop_stats = pd.read_csv('../../data/zaf_admpop_adm2_2020.csv')
print(pop_stats.shape)
pop_stats.sample(n=10)
```

```python
# Remove NaN data from pop_stats

pop_stats2 = pop_stats.loc[pop_stats['ADM0_EN'].notna()]
print(pop_stats2.shape)
print(pop_stats2.columns)
pop_stats2.sample(n=5)
```

```python
# remove unwanted columns

pop_stats3 = pop_stats2.drop(columns=['ADM0_EN', 'ADM0_PCODE', 'Unnamed: 61', 'Unnamed: 62', 'Unnamed: 63', 'Unnamed: 64',
       'Unnamed: 65', 'Unnamed: 66', 'Unnamed: 67', 'Unnamed: 68',
       'Unnamed: 69', 'Unnamed: 70', 'Unnamed: 71'])

print(pop_stats3.shape)
print(pop_stats3.columns)
pop_stats3.sample(n=5)
```

```python
# Add common district name and district code

unique_districts = sa_districts['DISTRICT.1'].unique()
```

```python
import numpy as np
from fuzzywuzzy import process, fuzz

adm2_name_map_district1 = {}
poor_results = []

for index, row in pop_stats3.iterrows():
    search = process.extract(row['ADM2_NAME'], unique_districts, scorer=fuzz.token_set_ratio)

    if search[0][1] >=75 :
        adm2_name_map_district1[row['ADM2_NAME']] = search[0][0]
    else:
        poor_results.append({'target': row['ADM2_NAME'], 'res': search})

print(len(adm2_name_map_district1))
print(len(poor_results))

for poor_result in poor_results:
    print(poor_result['target'], poor_result['res'][0])

# for i, key in enumerate(adm2_name_map_district1):
#     print(key, '-->', adm2_name_map_district1[key])
```

```python
def keycheck(search):
    print(search[0][0] in adm2_name_map_district1)


# Siyanda via google
search = process.extract('ZF Mgcawu', unique_districts, scorer=fuzz.token_set_ratio)
print(search)
keycheck(search)

# CACADU
search = process.extract('Sarah Baartman District', unique_districts, scorer=fuzz.token_set_ratio)
print(search)
keycheck(search)

# SISONKE
search = process.extract('Harry Gwala', unique_districts, scorer=fuzz.token_set_ratio)
print(search)
keycheck(search)

# UGO
search = process.extract('Ugu', unique_districts, scorer=fuzz.token_set_ratio)
print(search)
keycheck(search)

# UMHLABUYALINGANA
search = process.extract('uMkhanyakude', unique_districts, scorer=fuzz.token_set_ratio)
print(search)
keycheck(search)
```

```python
# Update these values in the map

adm2_name_map_district1['UGO'] = 'Ugu'
adm2_name_map_district1['SISONKE'] = 'Harry Gwala'
adm2_name_map_district1['UMHLABUYALINGANA'] = 'Umkhanyakude'
adm2_name_map_district1['CACADU'] = 'Sarah Baartman'
adm2_name_map_district1['SIYANDA'] = 'Z F Mgcawu'
```

```python
pop_stats4 = pd.DataFrame()
pop_stats4 =  pop_stats3
pop_stats4['DISTRICT_NAME'] = pop_stats3['ADM2_NAME'].map(adm2_name_map_district1)

len(pop_stats4['DISTRICT_NAME'].unique())
pop_stats4.sample(n=5)
```

```python
# Add the district code

pop_stats5 = pop_stats4
pop_stats5["DISTRICT_CODE"] = pop_stats4['DISTRICT_NAME'].map(dict(zip(sa_districts["DISTRICT.1"], sa_districts["DISTRICT"])))


print(len(pop_stats5['DISTRICT_CODE'].unique()))
pop_stats5.sample(n=10)
```

```python
pop_stats5.to_csv('../../data/cleaned/sa_district_popdem.csv')
```
