```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
data_location = "../../data/health_facilities_by_districts_regions.csv"
```

```python
data = pd.read_csv(data_location)
```

```python
print(data.shape)
data
```

It appears that the data is in csv and the geometry data for the District is stored in the polygon column in geojson format

```python
sa_municipal_boundaries_loc = "../../data/MDB_Local_Municipal_Boundary_2018.geojson"
sa_municipal = gpd.read_file(sa_municipal_boundaries_loc)
```

```python
print(sa_municipal.shape)
sa_municipal
```

```python
# Find distinct districts by district (the code)

districts = sa_municipal['DISTRICT'].unique()
print("Number of unique districts:", len(districts))
districts
```

There are the same amount of districs in both databases, combine the polygons from the one with the data from the other:

```python
geometries  = []

for district in data["DISTRICT"]:
    # get all municipalites from the required district
    municipalities = sa_municipal.loc[sa_municipal["DISTRICT"] == district]
    # Merge municipalies geometries to create district border:
    geom = municipalities.unary_union
    geometries.append(geom)

geometries
```

```python
new_data = data.drop(columns=['geometry'])

new_data = gpd.GeoDataFrame(new_data, crs=sa_municipal.crs, geometry=geometries)
new_data.head(5)
```

```python
new_data.plot(column="Population", legend=True)
new_data.plot(column='No. Public Hospitals', legend=True)
```
